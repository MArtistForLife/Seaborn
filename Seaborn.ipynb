#### Data Visualization With Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just a few lines of code. In this lesson, we will cover the basics of Seaborn, including creating various types of plots and customizing them.

```python
import seaborn as sns
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
##create a scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data=tips) # data = name of df you're using
plt.title("Scatter Plot of Total Bill vs. Tip")
plt.show()
```

```python
## Line Plot

sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line Plot of Total bill by Size")
plt.show()
```

```python
## Categorical Plots
## Bar Plot
sns.barplot(x='day',y='total_bill',data=tips)
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
## Box Plot
sns.boxplot(x="day",y='total_bill',data=tips) # yay!! box-and-whisker plots!!
plt.title('Box-and-Whisker Plot of Bill by Day')
plt.show()
```

```python
## Violin Plot...lol not violent /s

sns.violinplot(x='day',y='total_bill',data=tips)
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
# indicate which column from tips df to use (put in brackets)
# bins is how many bars you want (more = precise)
# kernel density estimator = gives you a curve
plt.title('Histogram of Total Bill by Count')
plt.show()
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
plt.title('KDE Plot of Total Bill by Density')
plt.show()
# just get that curve and the area under it
# use the column from tips df in brackets again
# fill shades in the area underneath
```

```python
# Pairplot
sns.pairplot(tips)
```

```python
## Heatmap
corr=tips[['total_bill','tip','size']].corr() # ooh cool :D
# use the columns you need from the tips df
corr
```

```python
sns.heatmap(corr,annot=True, cmap='coolwarm')
# annot = True will give the numbers inside the squares
# cmap = 'coolwarm' will give you a blue/red gradient, otherwise you get tan, magenta, purple, and black
# if correlation is really high, it'll be redder; if correlation is really low, bluer
```
